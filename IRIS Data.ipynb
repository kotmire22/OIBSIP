### Machine Learning Algorithm & Model Building 

###### We have to Classify the flower type using flower feature so this is a Supervised Learning Classification Problem 

```python
from sklearn.model_selection import train_test_split
```

```python
x=flower[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y=flower["Species"]
```

```python
#Y=y.array.reshape(-2,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train,y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtclf = DecisionTreeClassifier(criterion='entropy')
    
```

```python
dtclf.fit(X_train,y_train)
```

```python
Y_pred=dtclf.predict(X_test)
```

```python
metrics.accuracy_score(y_test,Y_pred)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf.fit(X_train, y_train)
```

```python
import graphviz 
from sklearn.tree import export_graphviz
```

```python
dot_data = export_graphviz(clf, 
                         out_file=None, 
                         feature_names=flower.drop('Species',1).columns,  
                         class_names=flower["Species"].unique(),  
                          filled=True, 
                          rounded=True,
                          special_characters= False)  
graph = graphviz.Source(dot_data) 
```
